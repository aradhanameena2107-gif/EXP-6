```python
# 6.5 get the input from the user of 3 subjects and calculate the gradecalculate average  grade using if-elif-else

a = int(input("Enter subject 1 marks: "))
b = int(input("Enter subject 2 marks: "))
c = int(input("Enter subject 3 marks: "))

avg = (a + b + c) / 3

if avg >= 60:
    print("Grade: A")
elif avg >= 50:
    print("Grade: B")
elif avg >= 40:
    print("Grade: C")
else:
    print("Grade: Fail")
```

```python
# 6.8 Salary Calculation with Allowance
# Problem: Write a Python program to calculate the gross salary of a employee
# If basic <= 10000: HR = 20%, DA = 80%
# If basic <= 20000: HRA = 25%, DA = 90%
# Otherwis: HRA = 30%, DA = 95%

basic = float(input("Enter basic salary: "))

if basic <= 10000:
    hra = basic * 0.20
    da = basic * 0.80
elif basic <= 20000:
  hra = basic * 0.25
  da = basic * 0.90
else:
  hra = basic * 0.30
  da = basic * 0.95
gross = basic + hra + da
print("Gross Salary:", gross)
```

```python
# 6.9 Write a Python program to calculate income tax based on annual income:
# Up to Rs.2,50,000: No tax
# Rs.2,50,001 - Rs.5,00,000: 5%
# Rs.5,00,001 - Rs.10,00,000: 20%
# Above Rs.10,00,000: 30%

income = float(input("Enter annual income: "))

if income <= 250000:
    tax = 0
elif income <= 500000:
    tax = (income - 250000) * 0.05
elif income <= 1000000:
    tax = (250000 * 0.05) + (income - 500000) * 0.20
else:
    tax = (250000 * 0.05) + (500000 * 0.20) + (income - 1000000) * 0.30

print("Income Tax:", tax)
```
